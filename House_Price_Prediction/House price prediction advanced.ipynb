```python
train = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
test = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

```python
train_data = train.copy()
```

```python

missing = [x for x in train_data.columns if train_data[x].isnull().any()]

print(f"Columns with missing value are :\n{missing}")
```

```python
def add_features(df):
    df = df.copy()
    # Total square footage
    df['TotalSF'] = df['GrLivArea'] + df['TotalBsmtSF'].fillna(0) + df['1stFlrSF'] + df['2ndFlrSF']
    # House age at sale
    df['HouseAge'] = df['YrSold'] - df['YearBuilt']
    # Remodel age
    
    df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
    return df
```

```python
train_data.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = train_data.SalePrice
train_data.drop(['SalePrice'], axis=1, inplace=True)
```

```python
# Cardinality = the number of unique values in a column
# Select categorical columns with  low cardinality 
categorical_cols = [cname for cname in train_data.columns if
                    train_data[cname].nunique() < 11 and 
                    train_data[cname].dtype == "object"]

print(f"Columns with categorical values are :\n{categorical_cols}")
```

```python
numerical_cols = [cname for cname in train_data.columns if 
                train_data[cname].dtype in ['int64', 'float64']]

print(f"Columns with numerical values are :\n{numerical_cols}")
```

```python
from sklearn.preprocessing import OrdinalEncoder

num_x_tr = train_data.copy()
num_x_tst = test.copy()

ord_enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

num_x_tr[categorical_cols] =  ord_enc.fit_transform(train_data[categorical_cols])
num_x_tst[categorical_cols] =  ord_enc.transform(test[categorical_cols])
```

```python
highCardinality_cols = [cname for cname in num_x_tr.columns if
                    num_x_tr[cname].nunique() > 10 and 
                    num_x_tr[cname].dtype == "object"]
print(highCardinality_cols)
```

```python
num_x_tr.drop(highCardinality_cols,axis=1,inplace=True)
num_x_tst.drop(highCardinality_cols,axis=1,inplace=True)
num_x_tst.info()
```

```python


missing = [i for i in num_x_tr.columns if num_x_tr[i].isnull().any()]
print(f"Columns with missing values = {missing}")
```

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='mean')
imp_x_train = pd.DataFrame(my_imputer.fit_transform(num_x_tr))
imp_x_test = pd.DataFrame(my_imputer.transform(num_x_tst))

imp_x_train.columns = num_x_tr.columns
imp_x_test.columns = num_x_tst.columns
```

```python
# my_cols = categorical_cols + numerical_cols
X_train = imp_x_train.copy()
X_test = imp_x_test.copy()

print(f"Training data: {X_train.shape}")
print(f"testing data : {X_test.shape}")
```

```python
from sklearn.model_selection import train_test_split

trainX,valX,trainY,valY = train_test_split(X_train,y,random_state=1,train_size=0.9)
print(f"Training data: {trainX.shape}")
print(f"testing data : {valX.shape}")
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=700,max_depth=4,learning_rate=0.04, random_state=1)
model.fit(trainX,trainY)
```

```python

validation = model.predict(valX)
```

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, confusion_matrix, classification_report



# MAE (mean absolute error) — just for consistency with your previous code
print("Training MAE:", mean_absolute_error(valY, validation))

```
